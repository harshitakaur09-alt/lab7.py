```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/Dataset for Data Analytics - Sheet1.csv')

df.head()
```

```python
print(df.shape)

df.info()

df.describe()

df.isnull().sum()
```

```python
df['CouponCode'] = df['CouponCode'].fillna(
    df['CouponCode'].mode()[0]
)

df.isnull().sum()
```

```python
Q1 = df['TotalPrice'].quantile(0.25)
Q3 = df['TotalPrice'].quantile(0.75)

IQR = Q3 - Q1

lower = Q1 - 1.5*IQR
upper = Q3 + 1.5*IQR

outliers = df[
(df['TotalPrice']<lower) |
(df['TotalPrice']>upper)
]

print(outliers.shape)
```

```python
df = df[
(df['TotalPrice']>=lower) &
(df['TotalPrice']<=upper)
]
```

```python
df['ProductValue'] = (
df['Quantity'] *
df['UnitPrice']
)
```

```python
df['CouponUsed'] = (
df['CouponCode']!='No Coupon'
).astype(int)
```

```python
df['OrderSize']=(
df['ItemsInCart']
.apply(
lambda x:
'Small'
if x<=2
else 'Large'
)
)
```

```python
df['Date'] = pd.to_datetime(
    df['Date'],
    format='%d-%m-%Y'
)

df['Month'] = df['Date'].dt.month

df['Year'] = df['Date'].dt.year
```

```python
df[['Date','Month','Year']].head()
```

```python
df['Date']=pd.to_datetime(
    df['Date'],
    dayfirst=True,
    errors='coerce'
)
```

```python
df['Date'].isnull().sum()
```

```python
df['PaymentMethod'].value_counts().plot(
kind='bar'
)

plt.title("Payment Method Distribution")

plt.show()
```

```python
plt.hist(
df['TotalPrice']
)

plt.title(
"Total Price Distribution"
)

plt.show()
```

```python
df.to_csv(
'cleaned_dataset.csv',
index=False
)

print(
"Dataset Saved Successfully"
)
```
